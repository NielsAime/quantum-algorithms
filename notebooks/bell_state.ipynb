```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Créer le circuit Bell state (|00> - |11>) / sqrt(2)
qc = QuantumCircuit(2, 2)
qc.h(0)       # Hadamard sur qubit 0 → superposition
qc.cx(0, 1)   # CNOT → entanglement
qc.z(1)       # Z gate pour avoir |00> - |11> et pas |00> + |11>
qc.measure([0, 1], [0, 1])

# Afficher le circuit
qc.draw('mpl')
```

```python
# Lancer la simulation
sim = AerSimulator()
result_ideal = sim.run(qc, shots=2048).result()
counts = result_ideal.get_counts()
plot_histogram(counts)
```
